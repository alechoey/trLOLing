* ## Introduction
We built a data product for the popular computer game, League of Legends, a multiplayer online game, where players control an individual champion and try to destroy the enemy base by killing enemy team members and completing objectives. Since champions and their unique abilities play an important role in the outcome of the game, we built a product to help answer the question of which set of champion selections are most likely to lead to victory. We will specifically apply this to ranked, draft-pick play, a mode of gameplay where each team begins by banning three champions so no member on either team may select them in the match. After the banning phase, each team alternates having the next team member select a champion to play. Therefore, given the list of banned champions and the current draft picks of both the friendly and enemy team, our product will suggest a champion to select next that will maximize the probability of winning, given data of previous matches.

```python
import math
variables = {'team1goldSpentNormalized': 1.0,
             'team2goldSpentNormalized': -1.0,
             'team1goldEarnedNormalized': 1.0,
             'team2goldEarnedNormalized': -1.0,
             'team1totalTimeCrowdControlDealtNormalized': 1.0,
             'team2totalTimeCrowdControlDealtNormalized': -1.0,
             'team1wardPlacedNormalized': 1.0,
             'team2wardPlacedNormalized': -1.0,
             'team1minionsKilledNormalized': 1.0,
             'team2minionsKilledNormalized': -1.0,
             'team1championsKilledNormalized': 1.0,
             'team2championsKilledNormalized': -1.0}
# Finally, export to edge file
def calculate_weight(row):
    score = int(sum(map(lambda key: variables[key] * (0 if math.isnan(row[key]) else row[key]), variables)) * 10000)
    return score if row['team1win'] else -score

game_champs['Source'] = game_champs.apply(lambda row: int(row['team1vertexId']) if row['team1win'] else int(row['team2vertexId']), axis=1) 
game_champs['Target'] = game_champs.apply(lambda row: int(row['team2vertexId']) if row['team1win'] else int(row['team1vertexId']), axis=1)
print game_champs[['Source', 'Target']].head(10)

game_champs_with_info['Weight'] = game_champs_with_info.apply(calculate_weight, axis=1)
game_champs_with_info['Source'] = game_champs_with_info.apply(lambda row: int(row['team1vertexId']) if row['team1win'] else int(row['team2vertexId']), axis=1) 
game_champs_with_info['Target'] = game_champs_with_info.apply(lambda row: int(row['team2vertexId']) if row['team1win'] else int(row['team1vertexId']), axis=1)
print game_champs_with_info[['Source', 'Target', 'Weight']].head(10)
```
